```python
import copy
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Les états

```python
States=[]
Categories=['E1', 'E2', 'E3'] # les diff situation du joueur, tout comme leur probabilité d'occurence pE1, pE2, pE3

for i in range(5):
    for j in range(5):
        for q in range(4):
            for Eperception in Categories:
                
                if j >= 3-q and i+j < 8: # le joueur adverse a moins le nbre de challenge raté comme score, et le premier joueur a avoir 4 pts remporte le jeu
                    # done=False
                    
                    if i==4 or j==4 : # critere de fin de jeu
                        # done = True
                        Eperception=None # pas besoin
                    s= ((i, j), q, Eperception)
                    if s not in States:  # enelever les repetitions
                        States.append(s)
                        
```

L'environnement

```python
class Environnement():
    def __init__(self, pE1=0.3 , pE2=0.2, b=0.8, pE3=0.5, c=0.2):
        
        def proba( s, a, s_prime):
            ### 1e etape: calcul de proba pour passer au "score" de s_prime ###
            if s[2] == None : # l Eperception None est celle des etats terminaux
                return 0
                
            deltaS = [ s_prime[0][0] - s[0][0], s_prime[0][1] - s[0][1] ] # difference entre le score d'arrivé et départ
            deltaNbreChlge = s_prime[1] - s[1] # différence entre le nbr de challenge d'arrivé et départ
        
            if deltaS == [0,1]: # A perd le jeu 
             
                if s[2] == 'E1' : # E1 est un etat gagant à coup sûr
                    p = 0 
                elif a == 0 : # A n'a pas fait un challenge
                    
                    if deltaNbreChlge != 0 : # nbreChlge diminue sans que A ait contesté, ABSURDE
                        p = 0
                    else:
                        p = 1 # A perd a coup sûr 

                else: # A a fait un challenge
                    
                    if deltaNbreChlge == 0 : # A a raté un challenge et nbreChlge = cst, ABSURDE
                        p = 0
                    elif s[2]  == 'E2' :
                        p = 1-b
                    elif s[2]  == 'E3' : 
                        p = 1-c   
                    
            elif deltaS == [1,0] :# A gagne le jeu 
                    
                if s[2] == 'E1' : # E1 est un etat gagant à coup sûr 
                    p = 1     
                elif a == 0 : # A n'a pas fait un challenge
                    p = 0
                else:
                    if s[2]  == 'E2' :
                        p = b
                    elif s[2]  == 'E3':
                        p = c

            else: # transition impossible ( d'apres nos hypotheses, sinon, possible dans la réalité) # pas sur qu il y ait d autre, juste précaution
                p = 0

                ### 2e etape: multiplication par la proba de tomber dans l Eperception de s_prime ###
            
            if  s_prime[2]  == 'E1':# on met a jour p pour prendre en compte 'la Eperception' du prochain etat
                p*=pE1
            elif s_prime[2]  == 'E2':
                p*=pE2
            elif s_prime[2]  == 'E3':
                p*=pE3
            # sinon, le Eperception est None, donc etat terminal 
            
            return p

            ### FIN de la def de Proba ###
        
        ### Remplissage du dictionaire de transition ###

        P = {}       
        for s in States:
            i, j = s[0][0], s[0][1] # le score
            q = s[1] # challenges restants
            Eperception = s[2]
            nA = 2 # nombres d'actions disponibles 
            
            dynamics_s = {} # le dico qui contiendra les infos de transition de chaque etat s
             
            ### Remplissage de dynamics_s ###
            
            if q == 0 or Eperception == 'E1' : # A ne peut plus challenger ou bien il est dans un etat gagnant
                nA = 1
                
            for a in range(nA):
                margeQ1 = 0 # intret dans le boucle ci apres
                margeQ2 = 2
                if a == 0:
                    margeQ2=1
                else:
                    margeQ1=1
                    
                s_prime_list = [] # succeseurs de s

                if i < 4 and j < 4: # les etats terminaux n'ont pas de successeurs
                
                    for t in range(margeQ1,margeQ2): # le petit delta q, il doit etre nul si a==0
                        for Eperception in Categories:
                            
                            if i+1 < 4: 
                                s_prime1 = ((i+1, j), q, Eperception) # on ne change pas q car A gagne dans ici
                                done1 = False
                                reward1 = 0 
                            elif i+1 == 4: 
                                s_prime1 = ((i+1, j), q, None)
                                reward1 = 100 # etat terminal gagnant
                                done1 = True
                            p1 = proba(s, a, s_prime1)
                            
                            if j+1 < 4:
                                s_prime2 = ((i, j+1), q-t, Eperception)
                                done2 = False
                                reward2 = 0
                            elif j+1 == 4:
                                s_prime2 = ((i, j+1), q-t, None)
                                done2 = True
                                reward2 = -100
                            p2 = proba(s, a, s_prime2)
                            
                            if (p1, s_prime1, reward1, done1) not in s_prime_list:   # il y a repetition avec les Eperception None
                                s_prime_list.append((p1, s_prime1, reward1, done1))
                            if(p2, s_prime2, reward2, done2) not in s_prime_list:  
                                s_prime_list.append((p2, s_prime2, reward2, done2))

                dynamics_s.update({a:s_prime_list}) # attention, les etats terminaux n ont pas de succeseur

            ### FIN remplissage de dynamics_s ###
            
            P.update({s: dynamics_s})
            
        ### FIN Remplissage du dictionaire de transition ###     
        
        self.P = P  
```

Résolution

fonction de valeur d'état d'action

```python
def compute_q_value_for_s_a(env, V, s, a, gamma):
    q = 0
    for (p_sPrime, sPrime, r_ss_a, done) in env.P[s][a]:
        q += p_sPrime * (r_ss_a + gamma * V[sPrime])
    return q
```

Initialisation 

```python
env=Environnement()
pi = {state: [0.5,0.5] for state in States} 
V = {state: 0 for state in States} 

gamma = 0.99 #facteur de remise du return, j'aurai pu prendre gamma = 1 car States est fini
epsilon = 0.00001 #seuil de similitude requis pour stopper les updates
```

Policy iteration

```python
def evaluate_policy(env, pi, V, gamma, epsilon):
    V_updated = copy.deepcopy(V)
    improved = True
    
    while True:        
        delta = 0
        for s in States:
            V_new = 0
            nA = len(env.P[s])
            
            for a in range(nA):
                prob_a = pi[s][a]
                q_s_a = compute_q_value_for_s_a(env, V_updated, s, a, gamma)
            
                V_new += prob_a * q_s_a
        
            delta = max(delta, np.abs(V_new - V_updated[s]))
            V_updated[s] = V_new

        if(delta < epsilon):
            break
    
    if(np.array_equal(V, V_updated)):
        improved = False
    
    return V_updated, improved

def improve_policy(env, pi, V, gamma):    
    for s in States:
        nA = len(env.P[s])
        q_s = np.zeros([nA, 1])
        for a in range(nA):
            q_s[a] = compute_q_value_for_s_a(env, V, s, a, gamma)
        
        best_a = np.argmax(q_s)
        pi[s] = np.eye(nA)[best_a]
    
    return pi
```

```python
i = 0
while True:
    i+=1
    
    V, improved = evaluate_policy(env, pi, V, gamma, epsilon)
    pi = improve_policy(env, pi, V, gamma)
    
    if(improved == False):
        print("Terminé après " + str(i) + " itérations.")
        break
```

le résultat

```python
# Les données fournies
data = E3

# Préparer les combinaisons de score et le nombre de challenges restants
scores = sorted(set(key[0] for key in data.keys()))
challenges = sorted(set(key[1] for key in data.keys()))

# Créer une matrice pour stocker les décisions
decision_matrix = np.full((len(challenges), len(scores)), -1)  # Initialiser avec -1 pour les valeurs manquantes

# Remplir la matrice de décision
for key, value in data.items():
    score, challenges_left, _ = key
    decision = np.argmax(value)  # Obtenir l'indice de la décision (0 ou 1)
    score_idx = scores.index(score)
    challenges_idx = challenges.index(challenges_left)
    decision_matrix[challenges_idx, score_idx] = decision

# Créer les étiquettes pour les axes
score_labels = [f"{s[0]}-{s[1]}" for s in scores]

# Définir la palette de couleurs : gris pour les manquants, vert pour challenger (1)
cmap = sns.color_palette(["grey","red", "green"])

# Création de la heatmap
plt.figure(figsize=(14, 10))
ax = sns.heatmap(decision_matrix, annot=False, cbar=True, cmap=cmap, xticklabels=score_labels, yticklabels=challenges, linewidths=.5, linecolor='black')

# Personnalisation de la colorbar
colorbar = ax.collections[0].colorbar
colorbar.set_ticks([-1, 0, 1])
colorbar.set_ticklabels(['Manquant', 'Non-challenger', 'Challenger'])

plt.xlabel('ScoreA, ScoreB')
plt.ylabel('Nombre de challenges restants')
plt.title('Décision de challenger en fonction du score et du nombre de challenges restants')
plt.show()
```
